# Як новини про криптовалюти впливають на криптовалюти?

## Вступ

Потік інформації є одним із основних факторів, які допомагають визначати ціни на фінансові активи. Інвестори реагують на нову інформацію, коли вона надходить, тим самим впливаючи на ціни активів. Технології змінили не лише спосіб торгівлі на фінансових ринках, але й спосіб розповсюдження та аналізу фінансових новин. Провайдери новин пропонують інформацію трейдерам у всьому світі за лічені секунди. Новинні канали зв’язку складають значну частину потоку інформації в реальному часі, яку отримують трейдери. Крім того, ера Інтернету дозволяє легко отримати інформацію в соціальних мережах і мережевих веб-сайтах. Криптовалюта є одним із таких ринків, де потік інформації відіграє ключову роль у визначенні ціни.

Важко визначити змінні, які впливають на ціни криптовалюти, через неоднозначну природу криптовалюти. Як наслідок, на ціни криптовалют значною мірою впливають настрої інвесторів, уявлення, думки та потік інформації. Загальнодоступна ринкова інформація відіграє вирішальну роль у формуванні настроїв криптовалютних інвесторів і, далі, їх торгової поведінки. Недосвідчених інвесторів на цьому ринку легко переконати інформацією в соціальних мережах і зобразити стадну поведінку. Більшість цінових змін є результатом змін у ринковій інформації, що підтверджує ідею про те, що ринкові рухи є значним фактором непередбачуваності.

Naeem та ін. (2021b) надають докази взаємозалежності між настроями онлайн-інвесторів (за допомогою індексу щастя Twitter) і прибутковістю криптовалюти. Анаміка та ін. (2021) також показують, що настрої інвесторів (проксі-індекс Sentix) значно впливають на ціни криптовалюти. Далі це спостерігали Shahzad et al. (2022) про те, що твіти впливових особистостей, таких як Ілон Маск, вплинули на ціни біткойнів і доджкойнів. Цей неоднозначний характер і участь роздрібних інвесторів у криптовалютному ринку спонукають нас досліджувати вплив публічної інформації на криптовалюту. Зокрема, ми аналізуємо вплив настроїв, отриманих із заголовків новин, загальнодоступного джерела інформації, на активність ринку криптовалют.

  2. Price (example: 42,259.) - колонка містить ціну біткоїна на певний момент часу або за певний період. Це може бути ціна за один біткоїн у доларах США або в іншій валюті, в залежності від контексту та джерела даних.

## Попередня підготовка даних

Видалимо дані, які не пригодяться нам під час аналізу та передбаченню

```python
def clean_change(change):
    return float(str(change)[:-1])

bitcoin_df = bitcoin_df[['date', 'change']]
bitcoin_df['change'] = bitcoin_df['change'].apply(clean_change)

bitcoin_df.head()
```

```python
def get_sentiment_verdict(data):
    data = eval(data)
    return data['class']

crypto_news_df['sentiment'] = crypto_news_df['sentiment'].apply(get_sentiment_verdict)
```

* Створюємо послідовну модель нейронної мережі за допомогою tf.keras.Sequentialкласу з TensorFlow (Послідовна модель — це простий і поширений спосіб визначення нейронних мереж, у яких шари додаються послідовно).
* Додаємо перший шар до моделі. Це щільний шар (також званий повністю зв’язаним шаром) із 512 нейронами. Кожен нейрон цього шару отримує вхідні дані від усіх нейронів попереднього шару (або вхідні дані, якщо це перший шар).
* Додаємо до моделі шар вилучення. Випадання — це техніка регулярізації, яка випадковим чином встановлює певний відсоток активацій у шарі на нуль під час навчання. У цьому випадку 0.3 вказує на 30% відсіву. Це допомагає запобігти надмірному підбору моделі до навчальних даних, змушуючи її вивчати більш надійні функції.
* Додаємо ще один щільний шар із 128 нейронами. Подібно до першого щільного шару, він отримує дані від усіх нейронів попереднього шару (який у даному випадку є шаром вилучення)
* На останньому етапі додає кінцевий вихідний рівень моделі. Це ще один щільний шар із лише одним нейроном. Оскільки ви намагаєтеся передбачити одне безперервне значення (зміна у вашому випадку), мати один вихідний нейрон є доречним

Загалом, тут ми визначаємо нейронну мережу з двома прихованими щільними шарами (по 512 і 128 нейронів кожен) з активацією ReLU і шаром відсіву між ними. Кінцевий вихідний рівень має один нейрон для прогнозування постійної цільової змінної (зміни).

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

Задаємо розклад темпу навчання:
* initial_learning_rate=0.001: цей аргумент встановлює початкову швидкість навчання на 0,001.
* decay_steps=1000: цей аргумент визначає кількість кроків (епох або пакетів), після яких швидкість навчання зменшиться на коефіцієнт decay_rate. У цьому випадку швидкість навчання буде знижуватися кожні 1000 кроків.
* decay_rate=0.9: цей аргумент визначає швидкість розпаду. Тут швидкість навчання буде помножена на 0,9 після кожних 1000 кроків, фактично зменшуючи її на 10% щоразу.

Додаємо оптимізатор з розкладом швидкості навчання та складаємо модель.

```python
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=1000,
    decay_rate=0.9
)
optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)
model.compile(optimizer='adam', loss='mean_squared_error')
```

Запобігаємо overfitting моделі, зупиняючи процес навчання, коли продуктивність моделі в наборі перевірки перестає покращуватися(зупиняючи навчання до того, як воно почне запам’ятовувати шум або нерелевантні шаблони).

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

Навчаємо модель нейронної мережі на підготовлених навчальних даних, відстежуємо її продуктивність у наборі перевірки та потенційно припиняємо навчання раніше за допомогою зворотного early_stopping виклику, щоб запобігти overfitting.

```python
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_data=(X_test_scaled, y_test), callbacks=[early_stopping])
```

## Оцінка продуктивності моделі

loss: це значення, ймовірно, представляє втрату навчання вашої моделі. Втрата під час навчання — це MSE (середня квадратична помилка), розрахована на основі даних навчання під час процесу навчання. Менші втрати від навчання зазвичай вказують на те, що модель добре засвоїла шаблони в даних навчання.

Mean Squared Error on Test Set: значення представляє тестову MSE , яка є ключовим показником, який ми оцінювали.

У вашому випадку тестовий MSE трохи нижчий за втрату під час навчання, що є позитивним знаком. Це свідчить про те, що модель може не надто підлаштовуватися під тренувальні дані та має певну здатність узагальнювати невидимі дані (тестовий набір).
